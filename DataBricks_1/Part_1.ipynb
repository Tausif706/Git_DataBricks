```python
from pyspark.sql import Row

data = [
    Row(col1=i, col2=f"val{i}", col3=i*2, col4=float(i)/2, col5=bool(i%2))
    for i in range(10)
]
df = spark.createDataFrame(data)
display(df)
```

```python
df.createOrReplaceTempView("df")
```

```python
%sql
SELECT * FROM df 
```
