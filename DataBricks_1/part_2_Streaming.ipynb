```python
%sql
CREATE VOLUME workspace.default.streaming_checkpoints;
```

```python
streaming_df = spark.readStream \
    .format("rate") \
    .option("rowsPerSecond", 5) \
    .load()

# display(streaming_df, checkpointLocation = "/Volumes/workspace/default/streaming_checkpoints/rate_demo")
```

```python
from pyspark.sql.functions import *

transformed = streaming_df \
    .withColumn("doubled_value", col("value") * 2) \
    .withColumn("event_time", col("timestamp"))
```

```python
display(transformed, checkpointLocation = "/Volumes/workspace/default/streaming_checkpoints/rate_demo")
```

```python
streaming_query = transformed.writeStream \
    .format("delta") \
    .outputMode("append") \
    .option("checkpointLocation", "/Volumes/workspace/default/streaming_checkpoints/streaming_demo") \
    .trigger(once=True).table("workspace.default.streaming_events")
```

```python
streaming_query.stop()
```
